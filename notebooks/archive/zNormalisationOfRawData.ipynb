# Z-normalisation of raw data
simple notebook to z-normalise the raw data of our time series data of 2023 and 2024. This should be good for shape based clustering.

```python
from pathlib import Path
import numpy as np
import pandas as pd

INPUT_CSV = Path("../data/raw/sample_23.csv")
OUTPUT_DIR = Path("../data/processed")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

OUTPUT_CSV = OUTPUT_DIR / "sample_23_znorm.parquet"
```

```python
df = pd.read_csv(INPUT_CSV)

id_col = df.columns[0]
value_cols = df.columns[1:]

ids = df[[id_col]].copy()
X = df[value_cols].astype(float).copy()

def z_normalise_row(row, eps=1e-8):
    arr = row.to_numpy(dtype=float)
    mean = arr.mean()
    std = arr.std()
    if std < eps:
        return np.zeros_like(arr)
    return (arr - mean) / std

X_znorm = X.apply(z_normalise_row, axis=1, result_type="expand")
X_znorm.columns = value_cols

df_znorm = pd.concat([ids, X_znorm], axis=1)
display(df_znorm)
```

```python
df_znorm.to_parquet(OUTPUT_CSV, index=False)
print(f"Saved z-normalised dataset to: {OUTPUT_CSV.resolve()}")
```
