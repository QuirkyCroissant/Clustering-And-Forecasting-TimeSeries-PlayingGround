```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_2023 = "../data/raw/sample_23.csv"
file_2024 = "../data/raw/sample_24.csv"
```

```python
data_2023 = pd.read_csv(file_2023, index_col=0)
data_2024 = pd.read_csv(file_2024, index_col=0)
```

```python
display(data_2023)
```

```python
display(data_2024)
```

```python

plt.figure(figsize=(12, 6))

plt.plot(data_2023.columns, data_2023.iloc[0], label="2023 - Series 1")
plt.plot(data_2024.columns, data_2024.iloc[0], label="2024 - Series 1")

plt.xlabel("Day of Year")
plt.ylabel("Energy Consumption (kWh)")
plt.title("Energy Consumption Time Series")
plt.legend()
plt.show()
```

```python
num_series_to_plot = 20
num_cols = 5
num_rows = (num_series_to_plot + num_cols - 1) // num_cols

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows), sharex=True, sharey=True)
axes = axes.flatten()

for i in range(num_series_to_plot):
    if i < len(data_2023):
        axes[i].plot(data_2023.columns, data_2023.iloc[i], label=f"2023 - Series {i+1}")
        axes[i].plot(data_2024.columns, data_2024.iloc[i], label=f"2024 - Series {i+1}")
        axes[i].set_title(f"Series {i+1}")
        axes[i].legend()

for i in range(num_series_to_plot, len(axes)):
    fig.delaxes(axes[i])

fig.suptitle("Energy Consumption Time Series", fontsize=16)
plt.xlabel("Day of Year")
plt.ylabel("Energy Consumption (kWh)")
plt.tight_layout(rect=(0, 0, 1, 0.96))
plt.show()
```
